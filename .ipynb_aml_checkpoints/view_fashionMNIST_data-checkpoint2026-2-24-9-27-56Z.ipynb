```python
import matplotlib.pyplot as plt
import torch
import torchvision
import os
```

```python
data_dir = "./data/train.pt"

# Load the training data (ensure the path matches your local directory)
# Common location: './data/FashionMNIST/processed/training.pt'
#torch.serialization.add_safe_globals([torchvision.datasets.mnist.FashionMNIST])

dataset = torch.load(data_dir, weights_only=False)

print(f"Images Shape: {dataset.data.shape}")   # Expected: torch.Size([60000, 28, 28])
#print(f"Labels Shape: {targets.shape}") # Expected: torch.Size([60000])
```

```python
img, label = dataset[0]

# If you used a Transform (like ToTensor), you can check the shape:
print(img.shape) 
```

```python
# .squeeze() removes the channel dimension [1, 28, 28] -> [28, 28]
plt.imshow(img.squeeze(), cmap='gray') 
plt.title(f"Label: {label}")
plt.axis('off') # Hide the X and Y axes
plt.show()
```

```python
# 1. Define the human-readable class names
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat', 
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']

# 2. Set up the figure for 5 subplots in 1 row
plt.figure(figsize=(15, 3))

for i in range(5):
    # Extract the image and label tuple
    img, label = dataset[i]
    
    # 3. Add to subplot (1 row, 5 columns, current index i+1)
    plt.subplot(1, 5, i + 1)
    
    # img is (1, 28, 28), so we .squeeze() it to (28, 28) for plotting
    plt.imshow(img.squeeze(), cmap='gray')
    plt.title(class_names[label])
    plt.axis('off')

plt.show()
```
